```
import numpy as np
import matplotlib.pyplot as plt
import cv2
data = []
l = []
for i in sorted(os.listdir('/content/gdrive/My Drive/Training_Test/sketches_train')):    
    print(i)
    b = sorted(os.listdir(os.path.join('/content/gdrive/My Drive/Training_Test/sketches_train', i)))
    np.random.shuffle(b)
    for j in b[:100]:
        a = plt.imread('/content/gdrive/My Drive/Training_Test/sketches_train'+'/'+i+'/' +j)
        a = cv2.resize(a , (100,100))
        data.append(a)
        l.append(i)
        
```

```
data = np.asarray(data)
l = np.asarray(l)
np.save('/content/gdrive/My Drive/numpy arrays/sketches/sketchdata.npy', data)
np.save('/content/gdrive/My Drive/numpy arrays/sketches/sketchlabels.npy', l)
```
