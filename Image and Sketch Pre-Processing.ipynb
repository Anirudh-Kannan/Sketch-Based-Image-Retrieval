```
import os
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import cv2
datai = []
li = []
datas = []
ls = []
for i in sorted(os.listdir('/content/drive/My Drive/Training_Test/sketches_test/')):    
    print(i)
    for j in sorted(os.listdir(os.path.join('/content/drive/My Drive/Training_Test/sketches_test', i))):
        a = plt.imread('/content/drive/My Drive/Training_Test/sketches_test'+'/'+i+'/' +j)/255
        a = cv2.resize(a , (100,100))
        datai.append(a)
        li.append(i)
```
